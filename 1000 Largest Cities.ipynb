### 1000 Largest US Cities By Population - 2017 Population
- data from [Biggest Cities](https://www.biggestuscities.com/)

```python
from bs4 import BeautifulSoup
import requests
import csv
import os
```

```python
url = 'https://www.biggestuscities.com/'
```

```python
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', 'table-condensed')
```

```python
print(table.prettify())
```

```python
table.find('tr')
```

```python
for row in table.findAll('tr')[1:5]:
    col = row.findAll('td')
    rank = col[0].text
    city = col[1].find('a').text.strip()
    state = col[2].find('a').text
    population = col[3].text.strip().replace(',','')
    print(rank, city, state, population)
```

### Write the first line  

```python
fname = '1000_cities.csv'
```

```python
try: 
    os.remove(fname)
    print('Deleted {}!!!'.format(fname))
except FileNotFoundError:
    print('That file does not exist!!')
```

```python
with open('1000_cities.csv', mode='a') as thousand_cities:
    wr = csv.writer(thousand_cities, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    wr.writerow(['rank', 'city', 'state', 'population'])
```

```python
with open('1000_cities.csv', mode='r') as thousand_cities:
    read_row = csv.reader(thousand_cities, delimiter=',')
    for x in read_row:
        print(','.join(x))
```

```python
for row in table.findAll('tr')[1:]:
    col = row.findAll('td')
    rank = col[0].text
    city = col[1].find('a').text.strip()
    state = col[2].find('a').text
    population = col[3].text.strip().replace(',','')
    row = (rank, city, state, population)
    with open('1000_cities.csv', mode='a') as thousand_cities:
        wr = csv.writer(thousand_cities, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        wr.writerow(row)
```

```python
with open('1000_cities.csv', mode='r') as thousand_cities:
    read_row = csv.reader(thousand_cities, delimiter=',')
    for x in read_row:
        print(','.join(x))
```
